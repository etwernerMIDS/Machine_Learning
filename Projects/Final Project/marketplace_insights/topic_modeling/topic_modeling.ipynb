# Introduction

Topic modeling is an unsupervised machine learning technique that's capable of scanning a set of documents, detecting word and phrase patterns within them, and automatically clustering word groups and similar expressions that best characterize a set of documents. A "topic" consists of a cluster of words that frequently occur together. Using contextual clues, topic models can connect words with similar meanings and distinguish between uses of words with multiple meanings.

Like many unsupervised clustering algorithms, the number of clusters (in this case, topics) must be determined a priori. We will explore different methods of topic modeling using the latent Dirichlet distribution, and evaluate the goodness of fit of the topics for the corpora of reviews. 

## Literature

* [Topics in Semantic Representation](https://webfiles.uci.edu/msteyver/publications/Griffiths_Steyvers_Tenenbaum_2007.pdf)
* [Latent Dirichlet Allocation](http://jmlr.org/papers/volume3/blei03a/blei03a.pdf)

## The Dirichlet Distribution and Latent Allocation

A Dirichlet distribution is another type of probability distribution, but instead of sampling from events like a coin flip, it samples over another probability distribution. For the topic modeling, we assume that a document will contain words from more than one topic, but we also assume that a single document will have one main topic. The shape of this categorical distribution is what we expect the final Dirichlet distribution to be. 

```python
Image(filename='images/dirichlet_distribution.png', width=700)
```

To describe a topic, we put together the similar words from a single document. For instance, a review may have the words `coffee`, `strong`, `dark`, and `Keurig` that cluster closely together. This collection of words describes the topic, which is hidden in the corpus itself: in other words, the information is **latent**. 

The process of Latent Allocation Distribution, or LDA, will assume that word tokens in a document are usually about one topic and vice versa. Then, it will sample the words into random topics and see if the Dirichlet distribution holds. We repeat this process until no more tokens move between topics, and we have our result. 

## LDA Drawbacks

LDA does have a few drawbacks. In particular, longer documents can perform poorly, since with more vectors we may have poorer similarity values. However, as we've seen in EDA, most reviews in our dataset have under 80 words, which means that we can largely avoid this pitfall. 

In addition, word substrings may result in a false positive match, particularly if the words are tokenized individually (i.e.: "root beer" vs "root" and "beer"). We will tackle this with n-grams in a later section.

## Visualizing Topics

LDA will form the topics, but we still need to see the meaning of each topic. We can get the top 10 words most relevant words for each topic. 

```python
def get_topics(model, words, n=10):
    for i, topic in enumerate(model.components_):
        print(f"Topic {i}:")
        print()
        print(" ".join(
            [words[i] for i in topic.argsort()[: -n - 1: -1]]))
        print('\n')
    print()
```

```python
get_topics(lda_tfidf, tfidf_vectorizer.get_feature_names())
```

We can intuit what many of these topics may be. Topic 0, which has words like `cookie`, `cinnamon`, and `oreo`, is probably related to baking or desserts, Topic 3 is probably related to pet food, etc.

# Gensim Model With n-Grams

We can improve on the topic generation by using n-grams.

## Text Processing

```python
text = df['preprocessed_text'].tolist()
text = [t.split(' ') for t in text]
```

```python
bigram = gensim.models.Phrases(text, min_count=5, threshold=80)
trigram = gensim.models.Phrases(bigram[text], threshold=60)  

bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)
```

```python
print(trigram_mod[bigram_mod[text[3]]])
```

Here, we notice that instead of having the tokens split between `root` and `beer`, as well as `cough` and `medicine`, we keep this text together, which gives us more info. 

```python
text_trigrams = [trigram_mod[bigram_mod[review]] for review in text]
```

## Creating Corpus

Now we can re-create the corpus and fit the LDA model.

```python
# Create Dictionary
id2word = corpora.Dictionary(text_trigrams)

# Term Document Frequency
corpus = [id2word.doc2bow(review) for review in text_trigrams]

# View
print(corpus[:1])
```

```python
# Viewing the corpus in human readable format
[[(id2word[id], freq) for id, freq in cp] for cp in corpus[:1]]
```

## Building Topic Model

```python
lda_model = gensim.models.ldamodel.LdaModel(
    corpus=corpus,
    id2word=id2word,
    num_topics=10, 
    random_state=100,
    update_every=1,
    chunksize=100,
    passes=10,
    alpha='auto',
    per_word_topics=True
)
```

```python
with open("../model/lda_gensim_tri_grams.model", 'wb') as f:
    dill.dump(lda_model, f)
```

## Evaluating Topic Models

As LDA is an unsupervised method, we need a way to determine if the topics generated are accurate or useful, as well has have a way to compare between two different models. The simplest way is to have a human reviewer; humans with domain knowledge will be able to eyeball if the topics make sense. 

However, sometimes we need a more automated or quantitative method of evaluating topics, which is where the topic *coherence score* comes in.

Topic Coherence measures score a single topic by measuring the degree of semantic similarity between high scoring words in the topic. These measurements help distinguish between topics that are semantically interpretable and topics that have statistical significance between word vectors. 

A common measure of coherence is cosine similarity across the word vectors (`coherence='c_v'`), which is what we will use for evaluating topics. 

```python
CoherenceModel(model=lda_model, texts=text_trigrams, dictionary=id2word, coherence='c_v').get_coherence()
```

# Finding Best Number of Topics

As mentioned above, LDA does require the number of topics to be set a priori. We can tune this parameter by running multiple models with different values of `k`, and choosing the model with the best coherence score. 

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=2):
    coherence_values = []
    
    model_list = []
    
    for num_topics in range(start, limit, step):
        model = gensim.models.wrappers.LdaMallet(
            mallet_path, 
            corpus=corpus, 
            num_topics=num_topics, 
            id2word=dictionary
        )
        model_list.append(model)
        
        coherence_model = CoherenceModel(
            model=model, 
            texts=texts, 
            dictionary=dictionary, 
            coherence='c_v'
        )
        coherence_values.append(coherence_model.get_coherence())

    return model_list, coherence_values
```

Unfortunately, LDA, even with the MALLET wrapper, is a very slow process. Due to time and memory limitations, we will run the following on a subset of our data. We go through the same data cleanup and tokenization process as the previous section.

```python
df_small = resample(
    df, 
    replace=True,
    n_samples=100000,
    random_state=1,
    stratify=df['Score']
)
```

```python
bigram_small = gensim.models.Phrases(text_small, min_count=5, threshold=80)
trigram_small = gensim.models.Phrases(bigram_small[text_small], threshold=60)  

bigram_small_mod = gensim.models.phrases.Phraser(bigram_small)
trigram_small_mod = gensim.models.phrases.Phraser(trigram_small)
```

```python
text_trigrams_small = [trigram_small_mod[bigram_small_mod[review]] for review in text_small]
```

## Creating Corpus

```python
# Create Dictionary
id2word_small = corpora.Dictionary(text_trigrams_small)

# Term Document Frequency
corpus_small = [id2word_small.doc2bow(review) for review in text_trigrams_small]
```

## Finding Best Coherence Model

```python
model_list, coherence_values = compute_coherence_values(
    dictionary=id2word_small, 
    corpus=corpus_small, 
    texts=text_trigrams_small, 
    start=5,
    step=5,
    limit=60
)
```

```python
x = range(5, 60, 5)

plt.plot(x, coherence_values, label="Coherence Score")

plt.xlabel("Number of Topics")
plt.ylabel("LDA Mallet Coherence Score")

plt.legend(loc='best')

plt.show()
```

```python
best_model = model_list[np.array(coherence_values).argmax()]
best_num_topics = best_model.num_topics

print(f'The best LDA model is built with {best_num_topics} topics.')
```

By choosing the highest coherence score, we get a final result of 35 topics, with a coherence score of a little under 58% -- a marked improvement over our previous scores.

## Analysis of Topics for Positive and Negative Reviews

Here, we can see that the types of negative vs. positive topics can help the seller better understand their customers. The positive review topics look fairly standard. The customers and their dogs seem to like the product.

However, analyzing the negative review topics can show where the customers are unhappy. Two of the topics seem to be related to the durability of the toy, using words like `chewed`, `lasted`, `month_old`, and `year_old`. Another one of the topics seem to show that the size wasn't correct for their dog. And another topic implies that perhaps their dog wasn't interested in the toy. 

While not all of these insights from the negative reviews are actionable (i.e.: there's nothing the seller can do to encourage interest in the toy from the dog), there are still things that could help. For instance, they could provide more useful size information for the product to ensure that a customer doesn't accidentally purchase a wrong sized toy. Alternatively, future iterations of the toy could aim to make it more sturdy.
